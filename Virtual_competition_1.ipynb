**1.** Create a DataFrame `df` from this dictionary `data` which has the index `labels`.

````python
import pandas as pd
import numpy as np
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data, index=labels)
df
````

# **Diabetes Dataset EDA**

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
data = pd.read_csv('diabetes_model.csv')
data.head(10)
````

````python
data.corr()
````

1. Analyze individual variables (e.g., histograms, box plots).

## **Histogram**

````python
data.hist(figsize=(10, 10))
plt.show()
````

## **Boxplot**

````python
data.boxplot(figsize=(15, 5))
plt.show()
````

2. Explore relationships between pairs of variables (e.g., scatter plots, correlation matrices).

# **Scatter plot**

````python
plt.scatter(data['Age'], data['BMI'] )
plt.legend(['Age', 'BMI'])
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('Age')
plt.show()
````

# ***Corelation Matrices***

````python
data[['Age', 'BloodPressure']].corr()
````

3. Investigate interactions among multiple variables (e.g., pair plots, 3D plots).

````python
plt.figure(figsize=(10, 8))
sns.pairplot(data[['Age', 'BMI', 'Outcome']])
plt.show()
````
