```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop(columns=['No of Times Visited'], inplace=True)
```

```python
df.drop(columns=['ID','Latitude','Date House was Sold','Longitude'], inplace=True)
```

```python
df.isna().sum()
```

```python
numerical_columns = ['Sale Price', 'No of Bathrooms', 'Flat Area (in Sqft)', 'Lot Area (in Sqft)',
                     'Area of the House from Basement (in Sqft)', 'Zipcode',
                     'Living Area after Renovation (in Sqft)']
```

```python
plt.figure(figsize=(3, 4))
plt.show()
# Scatter plot 1: 'Area of the House from Basement (in Sqft)' vs 'Flat Area (in Sqft)'
plt.subplot(2, 2, 1)
sns.scatterplot(x=df['Flat Area (in Sqft)'], y=df['Area of the House from Basement (in Sqft)'])
plt.title('Area of the House from Basement (in Sqft) vs Flat Area (in Sqft)')

# Scatter plot 2: 'Flat Area (in Sqft)' vs 'Living Area after Renovation (in Sqft)'
plt.subplot(2, 2, 2)
sns.scatterplot(x=df['Flat Area (in Sqft)'], y=df['Living Area after Renovation (in Sqft)'])
plt.title('Flat Area (in Sqft) vs Living Area after Renovation (in Sqft)')

# Scatter plot 3: 'No of Bathrooms' vs 'Flat Area (in Sqft)'
plt.subplot(2, 2, 3)
sns.scatterplot(x=df['Flat Area (in Sqft)'], y=df['No of Bathrooms'])
plt.title('No of Bathrooms vs Flat Area (in Sqft)')

# Scatter plot 4: 'Flat Area (in Sqft)' vs 'Sale Price'
plt.subplot(2, 2, 4)
sns.scatterplot(x=df['Flat Area (in Sqft)'], y=df['Sale Price'])
plt.title('Flat Area (in Sqft) vs Sale Price')

plt.tight_layout()
plt.show()
```

```python
for column in df.select_dtypes(include=['number']):
  plt.figure()
  df.boxplot([column])
  plt.title(column)
```

```python
outlier_cols = []
for column in df.select_dtypes(include=['number']):
  q1 = df[column].quantile(0.25)
  q3 = df[column].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr
  outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
  count = outliers.shape[0]
  if count > 0:
    outlier_cols.append(column)
print(outlier_cols)
```

```python
for column in outlier_cols:
  q1 = df[column].quantile(0.25)
  q3 = df[column].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr
  df[column] = df[column].clip(lower_bound, upper_bound)
```

```python
import matplotlib.pyplot as plt
import math

# Select numeric columns from the DataFrame
numeric_columns = df.select_dtypes(include=['number']).columns

# Define the number of rows and columns for subplots
n_cols = 3  # Number of columns for the grid
n_rows = math.ceil(len(numeric_columns) / n_cols)  # Calculate the number of rows needed

# Create subplots with defined rows and columns
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(15, n_rows * 4))

axes = axes.flatten()
for i, column in enumerate(numeric_columns):
    df.boxplot(column, ax=axes[i])
    axes[i].set_title(column)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import OrdinalEncoder
product_type_order = ['No','Yes']
oe = OrdinalEncoder(categories=[product_type_order])
df['Waterfront View'] = oe.fit_transform(df[['Waterfront View']])
df.head()
```
